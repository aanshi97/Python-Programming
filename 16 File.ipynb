# 2) write files, append 

```python
f=open('a.txt','x')
```

```python
f=open('a.txt','a')                                            # append the data in a.txt file
print(f.write('now the file has more content'))       
f.close()
```

```python
f=open('a.txt','rt')                      # read data after appending in a.txt file
print(f.read())
```

```python
f=open('a.txt','w')                               # overrite the data ie write new data instead in old data
print(f.write('i have delete all the content'))
f.close()
```

```python
f=open('a.txt','rt')              #read the file after overrite
print(f.read())
f.close()
```
